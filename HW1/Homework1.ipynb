```python
#read the data file
import csv as csv 

#import programs to plot the data
import pandas as pd 
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#read the data files and store them as a preprocessed data file (df) and a raw data file (df2)
df = pd.read_csv('post-operative.data')
df2 = pd.read_csv('post-operative.data')
```

The next lines of code preprocessed the some of the raw data. The colums of L-CORE, L-SURF, L-BP,L-O2, SURF-STBL, CORE-STBL, BP-STBL were ordinal in character. This made it possible to replace the strings of high, mid, low with 3,2,1. The high label being ranked the highest of the numbers. This can also be done for L-O2, SURF-STBL, CORE-STBL, BP-STBL.

```python
#made column names for data to match the matrix
df.columns = ['L-CORE','L-SURF','L-O2','L-BP','SURF-STBL','CORE-STBL','BP-STBL','COMFORT','ADM-DECS']
df2.columns = ['L-CORE','L-SURF','L-O2','L-BP','SURF-STBL','CORE-STBL','BP-STBL','COMFORT','ADM-DECS']
```

```python
#plot the graphs in line with the text after the program runs
%matplotlib inline
```

This is the first step in preprocessing data with ordinal features. The text value is assigned to a numerical value based on its order. This part of the process is called mapping. 

```python
#preprocessing data that are ordinal features 
temp_mapping = {'high': 3,'mid': 2,'low': 1}
o2_mapping = {'excellent':4, 'good':3,'fair':2,'poor':1}
stable_mapping ={'stable':3,'mod-stable':2,'unstable':1}
```

The next step is to assign the column that the mapped numbers will replace. This is done by calling the column name from the data frame and attaching it to the correct reassignment value. In the next lines of code, the correct mapping values are assigned to the columns.

```python
df['L-CORE'] = df['L-CORE'].map(temp_mapping)
df['L-SURF'] = df['L-SURF'].map(temp_mapping)
df['L-BP'] = df['L-BP'].map(temp_mapping)
df['L-O2'] = df['L-O2'].map(o2_mapping)
df['SURF-STBL'] = df['SURF-STBL'].map(stable_mapping)
df['CORE-STBL'] = df['CORE-STBL'].map(stable_mapping)
df['BP-STBL'] = df['BP-STBL'].map(stable_mapping)
```
